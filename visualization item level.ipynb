## Supply, Demand and SR Distribution

```python
fig, axs = plt.subplots(1, 3, figsize=(16, 4))
axs[0].hist(data=df_reduced, x="Total Supply", color="Red")
axs[0].set_title("Total Supply Distribution")
axs[0].set(xlim=(0, 40))
axs[1].hist(data=df_reduced, x="Meals Saved", color= "Orange")
axs[1].set_title("Meals Saved Distribution")
axs[2].hist(data=df_reduced[df_reduced["SR"]<=1], x="SR", color= "Yellow")
axs[2].set_title("Saved Ratio Distribution")
axs[2].set(xlim=(0, 1))
```

## Supply and demand Distribution per operation typer per language

```python
#Supply and demand and SR per operation type per language
plt.figure(figsize=(16, 4))
b = sns.boxplot(data = df_reduced[df_reduced["Region Language"] == "french"], x="Total Supply", y="Store Category")
b.set(xlim=(None, 28))
b.set_title("Supply per operation type in CH FR");

plt.figure(figsize=(16, 4))
c = sns.boxplot(data = df_reduced[df_reduced["Region Language"] == "german"], x="Total Supply", y="Store Category")
c.set(xlim=(None, 40))
c.set_title("Supply per operation type in CH DE");

plt.figure(figsize=(16, 4))
d = sns.boxplot(data = df_reduced[df_reduced["Region Language"] == "french"], x="Meals Saved", y="Store Category")
d.set(xlim=(None, 40))
d.set_title("Meals Saved per operation type in CH FR");

plt.figure(figsize=(16, 4))
e = sns.boxplot(data = df_reduced[df_reduced["Region Language"] == "german"], x="Meals Saved", y="Store Category")
e.set(xlim=(None, 40))
e.set_title("Meals Saved per operation type in CH DE");
```
