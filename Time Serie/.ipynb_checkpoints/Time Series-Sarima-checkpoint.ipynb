# Model Prep SARIMAX

## Functions

```python
# We define here a "Plot forecast vs. real", which also shows historical train set
def plot_forecast(forecast, train, test, upper=None, lower=None):
    is_confidence_int = isinstance(upper, np.ndarray) and isinstance(lower, np.ndarray)
    # Prepare plot series
    forecast_series = pd.Series(forecast, index=test.index)
    lower_series = pd.Series(upper, index=test.index) if is_confidence_int else None
    upper_series = pd.Series(lower, index=test.index) if is_confidence_int else None

    # Plot
    plt.figure(figsize=(10,4), dpi=100)
    plt.plot(train, label='training', color='black')
    plt.plot(test, label='actual', color='black', ls='--')
    plt.plot(forecast, label='forecast', color='orange')
    if is_confidence_int:
        plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.15)
    plt.title('Forecast vs Actuals')
    plt.legend(loc='upper left', fontsize=8);
```

```python
#evaluate
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - np.array(actual))/np.abs(np.array(actual)))  # MAPE
    me = np.mean(forecast - np.array(actual))             # ME
    mae = np.mean(np.abs(forecast - np.array(actual)))    # MAE
    mpe = np.mean((forecast - np.array(actual))/np.array(actual))   # MPE
    rmse = np.mean((forecast - np.array(actual))**2)**.5  # RMSE
    corr = np.corrcoef(forecast, np.array(actual))[0,1]   # corr
                    
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 
            'corr':corr})
```

## New Registered Users

```python
#Decomposition - linearizing
fig, axs = plt.subplots(2, 2, figsize=(18,8))
# keeping just log transform to stay ~ linear
gmr_NRU['linearized'] = np.log(gmr_NRU['NewRegisteredUsers'])

# linearized series
axs[0,0].plot(gmr_NRU.linearized); axs[0,0].set_title('linearized Series')

# Normal differencing
axs[0,1].plot(gmr_NRU.linearized.diff(1)); axs[0,1].set_title('1st Order Differencing')

# Seasonal differencing
axs[1,0].plot(gmr_NRU.linearized.diff(12))
axs[1,0].set_title('Seasonal differencing of period 12')

# Sesonal + Normal differencing
axs[1,1].plot(gmr_NRU.linearized.diff(12).diff(1))
axs[1,1].set_title('First order diff of seasonal differencing 12');
```

```python
from statsmodels.tsa import adfuller
# check with ADF Test for stationarity
print('p-value linearized Series: ', adfuller(gmr_NRU['linearized'].dropna())[1])
print('p-value 1st Order Differencing: ', adfuller(gmr_NRU['linearized'].diff(1).dropna())[1])
print('p-value Seasonal differencing of period 12: ', adfuller(gmr_NRU['linearized'].diff(12).dropna())[1])
print("p-value First order diff of seasonal differencing 12:", adfuller(gmr_NRU.linearized.diff(12).diff(1).dropna())[1])
```

```python
# Create a correct Training/Test split
train_size = 0.7
index_NRU = round(train_size*gmr_NRU.shape[0])
NRU_train = gmr_NRU.iloc[:index_NRU]
NRU_test = gmr_NRU.iloc[index_NRU:]
```

```python
#GridSearch
smodel = pm.auto_arima(NRU_train.linearized, seasonal=True, m=12, 
                       start_p=0, max_p=1, max_d=1, start_q=0, max_q=1,
                       start_P=0, max_P=2, max_D=1, start_Q=0, max_Q=2, 
                       trace=True, error_action='ignore', suppress_warnings=True)
```

```python
# Build Model
sarima_NRU = SARIMAX(NRU_train.linearized, order=(1, 0, 1), seasonal_order=(0, 0, 1, 12))
sarima_NRU = sarima_NRU.fit(maxiter=75)

# Forecast
NRU_results = sarima_NRU.get_forecast(len(NRU_test), alpha=0.05)
NRU_forecast = NRU_results.predicted_mean
NRU_confidence_int = NRU_results.conf_int()
```

```python
# Reconstruct by taking exponential
forecast_recons_NRU = pd.Series(np.exp(NRU_forecast), index=NRU_test.index)
#lower_recons = np.exp(NRU_confidence_int['lower linearized']).values
lower_recons_NRU = np.exp(NRU_confidence_int['lower linearized'])
upper_recons_NRU = np.exp(NRU_confidence_int['upper linearized'])

plot_forecast(forecast_recons_NRU, np.exp(NRU_train.linearized), np.exp(NRU_test.linearized), upper = upper_recons_NRU, lower=lower_recons_NRU)
```

```python
sarima_NRU.summary()
```

```python
forecast_accuracy(forecast_recons_NRU, np.array(np.exp(NRU_test.linearized)))
```

```python
residuals_NRU = pd.DataFrame(sarima_NRU.resid)

fig, ax = plt.subplots(1,2, figsize=(16,3))
residuals_NRU.plot(title="Residuals", ax=ax[0])
residuals_NRU.plot(kind='kde', title='Density', ax=ax[1]);
```

## Metrics definition

<b>mape</b>: The <b>mean absolute percentage error</b> is a measure of how accurate a forecast system is. It measures this accuracy as a percentage.

<b>me</b>: The <b>mean error</b> shows the average of the difference between actual and forecasted values.

<b>mae</b>: The <b>mean absolute error</b> takes the sum of the absolute difference from actual to forecast and averages them. It is less sensitive to the occasional very large error because it does not square the errors in the calculation.

<b>mpe</b>: The <b>mean Percentage Error</b> shows the average of the percent difference between actual and forecasted values. Both the ME and MPE will help indicate whether the forecasts are biased to be disproportionately positive or negative.

<b>rmse</b>: The <b>root mean squared error</b> represents the sample standard deviation of the differences between predicted values and observed values.

<b>corr</b>: <b>Correlation</b> between the test set and the forecast.
